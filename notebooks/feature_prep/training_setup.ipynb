### training the model

```python
import sagemaker
import boto3
import pandas as pd
from sagemaker import get_execution_role
from sagemaker.amazon.amazon_estimator import get_image_uri
from sagemaker.xgboost.estimator import XGBoost
from sagemaker.inputs import TrainingInput
```

```python
# Initialize variables
bucket_name = "sisyphus-general-bucket"
train_features_key = "AthenaInsights/data/model/train/train_features.parquet"
train_labels_key = "AthenaInsights/data/model/train/train_labels.parquet"
s3_output_location = f"s3://{bucket_name}/AthenaInsights/data/model-artifacts"


# Define the S3 paths for the training features and labels
train_features_path = 's3://sisyphus-general-bucket/AthenaInsights/data/model/train/train_features.parquet'
train_labels_path = 's3://sisyphus-general-bucket/AthenaInsights/data/model/train/train_labels.parquet'
```

```python
# Setup SageMaker session and role
session = sagemaker.Session()
# role = get_execution_role()
role = 'AmazonSageMaker-ExecutionRole-20241111T132700'
```

```python
# initialize hyperparameters
# hyperparameters = {
#         "max_depth":"5",
#         "eta":"0.2",
#         "gamma":"4",
#         "min_child_weight":"6",
#         "subsample":"0.7",
#         "objective":"binary:logistic",
#         "num_round":50
#         }

hyperparameters = {
        'objective': 'multi:softmax',
        'num_class': 3,                  # Since you have three classes A, B, C
        'num_round': 100                 # Number of boosting rounds
    }
```

```python


xgboost_estimator = XGBoost(
    entry_point='train.py',             # Training script (needs to be created, see below)
    role=role,
    instance_count=1,
    instance_type='ml.m5.xlarge',       # Choose your instance type
    framework_version='1.0-1',          # Choose your XGBoost version
    output_path=s3_output_location,
    sagemaker_session=session,
    hyperparameters=hyperparameters
)
```

```python
# Set the input configuration for your training job
data_channels = {
    'train_features': TrainingInput(train_features_path, content_type='parquet'),
    'train_labels': TrainingInput(train_labels_path, content_type='parquet')
}

# Start the training job
xgboost_estimator.fit(data_channels)
```
